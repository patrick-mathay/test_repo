Verficamos si existen duplicados para cada columna en este caso nos interesaria ver que en primero instancia:
-   No existan duplicados en la columna 'business_id'


```python
for columna in df_business.columns:
    hay_duplicados = df_business[columna].duplicated().any()
    if hay_duplicados:
        cantidad_duplicados = df_business[columna].duplicated().sum()
        print(f"En la columna '{columna}' hay {cantidad_duplicados} duplicados.")
    else:
        print(f"No hay duplicados en la columna '{columna}'.")
```

Graficamos para hacer más facil su interpretación 

```python
duplicados_por_columna = {}

# Iteraramos sobre cada columna del df
for columna in df_business.columns:
    hay_duplicados = df_business[columna].duplicated().any()
    if hay_duplicados:
        # Obtener la cantidad de duplicados en la columna
        duplicados = df_business[columna].duplicated().sum()
        # Almacenar el nombre de la columna y la cantidad de duplicados en el diccionario
        duplicados_por_columna[columna] = duplicados

# Ordenamos el diccionario  de mayor a menor
duplicados_ordenados = sorted(duplicados_por_columna.items(), key=lambda x: x[1], reverse=True)

# Extraemos los nombres de las columnas y las cantidades de duplicados ordenadas
nombres_columnas_ordenados = [x[0] for x in duplicados_ordenados]
cantidad_duplicados_ordenados = [x[1] for x in duplicados_ordenados]

# Creamos el diagrama de barras ordenado de mayor a menor
plt.figure(figsize=(10, 6))
plt.barh(nombres_columnas_ordenados, cantidad_duplicados_ordenados, color='skyblue')
plt.xlabel('Cantidad de Duplicados')
plt.ylabel('Columnas')
plt.title('Cantidad de Duplicados por Columna (Ordenado de Mayor a Menor)')
plt.show()
```

```python
# Función para contar las ocurrencias de la palabra "Restaurants"
def contar_restaurants(cadena):
    if isinstance(cadena, str):
        # Usar expresión regular para buscar coincidencias parciales
        coincidencias = re.findall(r'\bRestaurants\b', cadena)
        # Contar las coincidencias encontradas
        return len(coincidencias)
    else:
        return 0

# Aplicar la función a la columna 'categories' para contar las ocurrencias de "Restaurants"
df_business['num_restaurants'] = df_business['categories'].apply(contar_restaurants)

# Calcular el total de ocurrencias de "Restaurants"
total_restaurants = df_business['num_restaurants'].sum()

# Mostrar el total de ocurrencias
print(f"Total de ocurrencias de 'Restaurants': {total_restaurants}")

# Crear el gráfico de barras
plt.figure(figsize=(8, 6))
plt.bar(['Restaurants', 'Otro'], [total_restaurants, len(df_business) - total_restaurants], color=['lightgreen', 'lightgrey'])
plt.xlabel('Categorías')
plt.ylabel('Frecuencia')
plt.title('Ocurrencias de Restaurantes en la Columna "categories"')
plt.show()
```

Se podria realizar la division en dos grandes rubros:
1. Rubro de Gastronomía = "Restaurants", "Food" , "Mexican", "Pizza" ,"Chinese" ,"Coffee & Tea"
2. Rubro de Belleza = "Beauty & Spas" , "Nail Salons" , "Hair Salons"

Rubro de Gastronomía

```python
# Función para contar las ocurrencias de las palabras "Restaurants", "Food" , "Mexican", "Pizza" ,"Chinese" ,"Coffee & Tea"
def contar_restaurants_food(cadena):
    if isinstance(cadena, str):
        # Usar expresión regular para buscar coincidencias parciales 
        coincidencias = re.findall(r'\b(Restaurants|Food|Mexican|Pizza|Chinese|Coffee & Tea)\b', cadena)
        # Contar las coincidencias encontradas
        return len(coincidencias)
    else:
        return 0

# Aplicar la función a la columna 'categories' para contar las ocurrencias de "Restaurants" y "Food"
df_business['num_restaurants_food'] = df_business['categories'].apply(contar_restaurants_food)

# Calcular el total de ocurrencias de "Restaurants" y "Food"
total_restaurants_food = df_business['num_restaurants_food'].sum()

# Mostrar el total de ocurrencias
print(f"'Total de ocurrencias del Rubro Gastronomía': {total_restaurants_food}")

# Crear el gráfico de barras
plt.figure(figsize=(8, 6))
plt.bar(['Gastronomía', 'Otro'], [total_restaurants_food, len(df_business) - total_restaurants_food], color=['lightgreen', 'lightgrey'])
plt.xlabel('Categorías')
plt.ylabel('Frecuencia')
plt.title('Ocurrencias del Rubro Gastronomía en la Columna "categories"')
plt.show()
```

Rubro de Belleza

```python

# Función para contar las ocurrencias de las palabras "Beauty & Spas" , "Nail Salons" , "Hair Salons"
def contar_restaurants_food(cadena):
    if isinstance(cadena, str):
        # Usar expresión regular para buscar coincidencias parciales
        coincidencias = re.findall(r'\b(Beauty & Spas|Nail Salons|Hair Salons)\b', cadena)
        # Contar las coincidencias encontradas
        return len(coincidencias)
    else:
        return 0

# Aplicar la función a la columna 'categories' para contar las ocurrencias de "Beauty & Spas" , "Nail Salons" , "Hair Salons"
df_business['num_Salons'] = df_business['categories'].apply(contar_restaurants_food)

# Calcular el total de ocurrencias de "Beauty & Spas" , "Nail Salons" , "Hair Salons"
total_Salons = df_business['num_Salons'].sum()
print(f"'Total de ocurrencias del Rubro Belleza': {total_Salons}")

plt.figure(figsize=(8, 6))
plt.bar(['Belleza', 'Otro'], [total_Salons, len(df_business) - total_Salons], color=['skyblue', 'lightgrey'])
plt.xlabel('Categorías')
plt.ylabel('Frecuencia')
plt.title('Total de ocurrencias del Rubro Belleza en la Columna "categories"')
plt.show()
```

Ahora realizamos el compartivo de ambos rubros

```python
# Definir las palabras específicas a buscar
palabras_food = ["Restaurants", "Food" , "Mexican", "Pizza" ,"Chinese" ,"Coffee & Tea"]
palabras_beauty = ["Beauty & Spas", "Nail Salons", "Hair Salons"]

# Función para contar las ocurrencias de las palabras específicas
def contar_palabras_especificas(cadena, palabras_especificas):
    if isinstance(cadena, str):
        # Usar expresión regular para buscar coincidencias parciales
        coincidencias = re.findall(r'\b(?:' + '|'.join(palabras_especificas) + r')\b', cadena)
        # Contar las coincidencias encontradas
        return len(coincidencias)
    else:
        return 0

# Aplicar la función a la columna 'categories' para contar las ocurrencias de las palabras específicas
df_business['num_palabras_restaurants_food'] = df_business['categories'].apply(contar_palabras_especificas, palabras_especificas=palabras_food)
df_business['num_palabras_beauty_spas_nail_hair'] = df_business['categories'].apply(contar_palabras_especificas, palabras_especificas=palabras_beauty)

# Calcular el total de ocurrencias
total_restaurants_food = df_business['num_palabras_restaurants_food'].sum()
total_beauty_spas_nail_hair = df_business['num_palabras_beauty_spas_nail_hair'].sum()
otros = len(df_business) - total_restaurants_food - total_beauty_spas_nail_hair

# Mostrar el total de ocurrencias
print(f"Total de ocurrencias del Rubro Gastronomía': {total_restaurants_food}")
print(f"Total de ocurrencias del Rubro Belleza': {total_beauty_spas_nail_hair}")
print(f"Total de ocurrencias de 'Otros': {otros}")

# Crear el gráfico de barras
categorias = ['Gastronomía', 'Belleza', 'Otros']
frecuencias = [total_restaurants_food, total_beauty_spas_nail_hair, otros]
colores = ['lightgreen', 'skyblue', 'lightgrey']  # Lista de colores para cada barra

plt.figure(figsize=(8, 6))
plt.bar(categorias, frecuencias, color=colores)
plt.xlabel('Tipo de Categorías')
plt.ylabel('Frecuencia')
plt.title('Gastronomía vs Belleza "categories"')
plt.show()
```
