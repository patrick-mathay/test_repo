```python
#using bit manipulation aka shifting 
def NumeroBinario(numero):
    if type(numero) != int or numero < 0:
        return None
    binary_representation = 0
    power = 0
    while numero > 0:
        binary_representation += (numero % 2) * (10 ** power)
        numero //= 2
        power += 1
    return binary_representation

# Example:
resultado = NumeroBinario(10)
print(resultado)
```

```python
#using bin()
def NumeroBinario(numero):

# Verificar que el parámetro sea un número entero mayor o igual a cero
    if not isinstance(numero, int) or numero < 0:
        return None  # Retorna nulo si el parámetro no cumple con los requisitos

    # Convertir el número a su representación binaria y devolver el resultado como entero
    binario = int(bin(numero)[2:])
    return binario

# Ejemplos de uso:
# Caso válido
numero_entero = 10
resultado = NumeroBinario(numero_entero)
print(f"El número {numero_entero} en binario es: {resultado}")

# Caso inválido (no es un número entero)
numero_no_entero = 5.5
resultado_invalido = NumeroBinario(numero_no_entero)
print(f"Resultado para {numero_no_entero}: {resultado_invalido}")
```
